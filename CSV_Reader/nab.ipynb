```python
import pandas as pd
from pathlib import Path
```

```python
csvpath = Path("CSV_Reader/nab_3_months_Transactions_sample.csv")
df_nab = pd.read_csv(csvpath,parse_dates=True,infer_datetime_format=True, index_col='Date')
df_nab
```

```python
df_nab = df_nab.drop(columns = ['Account Number','Transaction Type','Balance','Category','Merchant Name'])

df_nab = df_nab.rename(columns={
    "Amount": "Transaction",
    "Transaction Details": "Detais"
})

df_nab
```

```python
df_nab.groupby([df_nab.index.year,df_nab.index.month]).sum()
```

```python
s = df_nab.index.isocalendar()
s
```

```python
df_by_week = df_nab.groupby(
    [s.year, 
     s.week]
).sum()

df_by_week
```
