```python
#Import libraries and dependencies
import pandas as pd
from pathlib import Path

#Set path for the data to be read
cba_path=Path("Resources/CBA.csv")

#Read data
cba_df=pd.read_csv(cba_path, header=None)

# Rewrite the column names
cba_df.head()
```

```python
columns = ["Date","Details","Account","Category","Transaction"]
cba_df.columns = columns
cba_df.head()
```

```python
cba_df=cba_df.drop(["Account","Category"], axis=1)
cba_df.head()
```

```python
cba_df=cba_df[["Date","Transaction","Details"]]
cba_df
```

```python
#Set date as index
cba_df.set_index(pd.to_datetime(cba_df['Date'], infer_datetime_format=True), inplace=True)

#Drop extra "Date" column
cba_df.drop(columns=['Date'], inplace=True)
cba_df.head()
```

```python
cba_df.dtypes
```

```python
cba_df.index
```

```python
cba_grp=cba_df.groupby([cba_df.index.year, cba_df.index.month]).sum()
cba_grp
```

```python
cba_grp_stat=cba_grp.describe()
print(cba_grp_stat)
```

```python
monthly_average=cba_grp_stat.iat[1,0]
```

```python
print(monthly_average)
```

```python
cba_grp1=cba_df.index.isocalendar()
cba_grp1
```

```python
cba_week=cba_df.groupby(
    [cba_grp1.year,
     cba_grp1.week]
).sum()

cba_week
```
